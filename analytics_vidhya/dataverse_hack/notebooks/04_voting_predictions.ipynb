```python
SUB_PATH = '/content/drive/MyDrive/data_science_competitions/analytics_vidhya/dataverse_hack/submissions'
folds_df = pd.DataFrame()
mode_df = pd.DataFrame()
```

```python
for i in range(1, 4):
    for j in range(1, 3):
        for k in range(5):
            path = f'{SUB_PATH}/nb_0{i}/0{j}_fold{k}.csv'
            col = f'nb0{i}_0{j}_fold{k}'
            folds_df[col] = pd.read_csv(path, usecols=[1])
```

```python
for i in range(1, 4):
    for j in range(1, 3):
        path = f'{SUB_PATH}/nb_0{i}/0{j}_mode.csv'
        col = f'nb0{i}_0{j}_mode'
        mode_df[col] = pd.read_csv(path, usecols=[1])
```

```python
folds_majority = round(folds_df.mean(axis=1)).astype('int') #1 only if more than half preds are 1
```

```python
mode_majority = round(mode_df.mean(axis=1)).astype('int')
```

```python
test_index = pd.read_csv(f'{SUB_PATH}/nb_01/01_mode.csv', usecols=[0])
```

```python
folds_voting = pd.DataFrame({
    'policy_id': test_index.policy_id.values,
    'is_claim': folds_majority.values
})
folds_voting.to_csv(f'{SUB_PATH}/nb_04/folds_voting.csv', index=False)
```

```python
mode_voting = pd.DataFrame({
    'policy_id': test_index.policy_id.values,
    'is_claim': mode_majority.values
})
mode_voting.to_csv(f'{SUB_PATH}/nb_04/mode_voting.csv', index=False)
```
